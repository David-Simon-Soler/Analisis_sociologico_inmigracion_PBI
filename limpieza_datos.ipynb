```python
import pandas as pd
import pyreadstat
```

```python
df, meta = pyreadstat.read_sav('3190.sav_BUENA.sav')
```

```python
print ("--- LAS CINCO PRIMERAS FILAS DEL DATASET ---")
print (df.head())
```

```python
print ("\n---DIMENSIONES DEL ARCHIVO ---") 
print (f"encuestados: {df.shape[0]} | Variables: {df.shape[1]}")
```

```python
print("\n--- DICCIONARIO DE VARIABLES (MUESTRA) ---")

# Fíjate bien: es .items() con la 's' al final y los paréntesis
for codigo, etiqueta in list(meta.column_names_to_labels.items())[:10]:
    print(f"{codigo} -> {etiqueta}")
```

```python
# ==========================================
# EL RASTREADOR DE VARIABLES (PARA NOTAS)
# ==========================================

# Buscamos en el 'meta' qué códigos corresponden a las preguntas que nos interesan
print("Buscando los códigos reales de tus variables")

# Palabras clave que queremos buscar en las preguntas
busqueda = ['EDAD', 'SEXO', 'SEGURIDAD', 'ESTUDIOS', 'POLICÍA', 'P13']

for codigo, pregunta in meta.column_names_to_labels.items():
    # Si alguna de nuestras palabras está en la pregunta (en mayúsculas)
    if any(p in pregunta.upper() for p in busqueda):
        print(f"CÓDIGO: {codigo} | PREGUNTA: {pregunta}")
```

```python
variables_reales = ['P12', 'P44', 'P45', 'P49A', 'CCAA']
df_reducido = df[variables_reales].copy()
```

```python
nombres_nuevos = {
    'P12': 'Inseguridad',
    'P44': 'Sexo',
    'P45': 'Edad',
    'P49A': 'Nivel_Estudios',
    'CCAA': 'Comunidad_Autonoma'
}

df_reducido = df_reducido.rename(columns=nombres_nuevos)

print("DATASET LIMPIO Y ORDENADO")
display(df_reducido.head())
```

```python
# 1. Volvemos a cargar el recorte pero CON LOS NOMBRES ORIGINALES
# (Así Python no se confunde al buscar las etiquetas en el 'meta')
variables_cis = ['P12', 'P44', 'P45', 'P49A', 'CCAA']
df_final = df[variables_cis].copy()

# 2. APLICAMOS LA TRADUCCIÓN (Convertir números a palabras)
# Lo hacemos aquí, cuando las columnas aún se llaman P12, P44, etc.
df_final = pyreadstat.set_value_labels(df_final, meta)

# 3. AHORA SÍ, RENOMBRAMOS PARA QUE SEA HUMANO
nombres_amigables = {
    'P12': 'Inseguridad',
    'P44': 'Sexo',
    'P45': 'Edad',
    'P49A': 'Nivel_Estudios',
    'CCAA': 'Comunidad_Autonoma'
}
df_final = df_final.rename(columns=nombres_amigables)

print("✅ ¡POR FIN! Datos traducidos y legibles:")
display(df_final.head())
```

```python
tabla_cruzada = pd.crosstab(df_final['Sexo'], df_final['Inseguridad'])

print ("--- TABLA CRUZADA ---")
display(tabla_cruzada)
```

```python
tabla_porcentajes = pd.crosstab (df_final['Sexo'], df_final['Inseguridad'], normalize='index') * 100

print ("\n--- PORCENTAJES POR SEXO (¿Quién siente más inseguridad?)")
display(tabla_porcentajes.round(2))
```

```python
busqueda_inmi = ['INMIGRA', 'TRABAJO', 'AYUDAS', 'CONVIVENCIA']
for codigo, pregunta in meta.column_names_to_labels.items():
    if any (p in pregunta.upper() for p in busqueda_inmi):
        print(f"CÓDIGO: {codigo} | PREGUNTA: {pregunta}")
```

```python
vars_interes = ['P44', 'P45', 'P49A', 'CCAA', 'P16', 'P21_4', 'P24_7', 'P23']
df_inmi = df[vars_interes].copy()
df_inmi = pyreadstat.set_value_labels(df_inmi, meta)
```

```python
nombres_inmi = {
    'P44': 'Sexo',
    'P45': 'Edad',
    'P49A': 'Estudios',
    'CCAA': 'Comunidad',
    'P16': 'Aportan_vs_Reciben',
    'P21_4': 'Quitan_Trabajo',
    'P24_7': 'Matrimonio_Hijo',
    'P23': 'Valoracion_General'
}
df_inmi = df_inmi.rename(columns=nombres_inmi)

print("Dataset de Inmigración")
display(df_inmi.head())
```

```python
# Tabla de contingencia en porcentajes
cruce_estudios_trabajo = pd.crosstab(df_inmi['Estudios'], 
                                     df_inmi['Quitan_Trabajo'], 
                                     normalize='index') * 100

print("¿QUIÉN CREE QUE QUITAN EL TRABAJO SEGÚN SUS ESTUDIOS?")
display(cruce_estudios_trabajo.round(2))
```

```python
cruce_sexo_matrimonio = pd.crosstab(df_inmi['Sexo'], 
                                    df_inmi['Matrimonio_Hijo'], 
                                    normalize='index') * 100
print ("DIFERENCIAS DE GÉNERO EN LA ACEPTACIÓN FAMILIAR")
display(cruce_sexo_matrimonio.round(2))
```

```python
bins = [18, 30, 50, 65, 100]
nombres_grupos = ['Jóvenes', 'Adultos', 'Adultos Mayores', 'Jubilados']
df_inmi ['Tramo_edad'] = pd.cut(df_inmi['Edad'], bins=bins, labels=nombres_grupos)
```

```python
cruce_edad_valoracion = pd.crosstab(df_inmi['Tramo_edad'], 
                                    df_inmi['Valoracion_General'], 
                                    normalize='index') * 100

print ("\n--- VALORACIÓN DE LOS INMIGRANTES POR EDAD")
display(cruce_edad_valoracion.round(2))
```

```python
# Guardamos la versión final "maestra"
df_inmi.to_csv('proyecto_inmigracion_final.csv', index=False, encoding='utf-8-sig')
```

```python
# ==========================================
# LIMPIEZA DE EDADES (VERSIÓN DEFINITIVA)
# ==========================================

# 1. CONVERSIÓN A NÚMERO (El paso que faltaba)
# 'errors=coerce' es magia: si encuentra algo que no es un número (como "N.S."), 
# lo convierte automáticamente en un nulo (NaN) sin dar error.
df_inmi['Edad'] = pd.to_numeric(df_inmi['Edad'], errors='coerce')

# 2. AHORA SÍ: Filtramos valores imposibles (como ese 470)
# Solo nos quedamos con valores entre 18 y 100
df_inmi.loc[df_inmi['Edad'] > 100, 'Edad'] = None

# 3. Eliminamos las filas que se hayan quedado vacías (nulos)
df_inmi = df_inmi.dropna(subset=['Edad'])

# 4. RECALCULAMOS LOS TRAMOS (Ahora que 'Edad' vuelve a ser numérica)
bins = [18, 30, 50, 65, 100]
labels = ['Jóvenes', 'Adultos', 'Adultos Mayores', 'Jubilados']
df_inmi['Tramo_edad'] = pd.cut(df_inmi['Edad'], bins=bins, labels=labels)

# 5. Guardamos el CSV final para Power BI
df_inmi.to_csv('proyecto_inmigracion_final.csv', index=False, encoding='utf-8-sig')

print(df_inmi['Edad'].describe()) # Verás que el Max ahora es normal
```
